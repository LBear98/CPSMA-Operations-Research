```python
def count_trailing_zeros(n):
    count = 0
    i = 5
    while n // i >= 1:
        count += n // i
        i *= 5
    return count

n = 1000
zeros = count_trailing_zeros(n)
print(f"The number of zeros at the end of {n}! is: {zeros}")
```

We also want to find where $f(x) = 3x + 2y$ is at its maximum. To do this we can substitute in the values of the corner points for $x$ and $y$, respectively. Like so:

$3(0) + 2(0) = 0$

$3(\frac{1}{2}) + 2(\frac{17}{2}) = \frac{37}{2}  18.5$

$3(0) + 2(9) = 18$

$3(\frac{11}{5}) + 2(0) = \frac{33}{5} \equiv 6.6$

We obtain our maximum when $x = \frac{1}{2}$ and $y = \frac{17}{2}$. Now let's show this through some code.

```python
import numpy as np

# Define the coefficients of the objective function and constraints
# Maximize f(x, y) = 3x + 2y
# Subject to:
# x + y <= 9
# 5x + y <= 11
# x >= 0, y >= 0

# Convert to standard form by adding slack variables s1 and s2
# Maximize f(x, y) = 3x + 2y + 0s1 + 0s2
# Subject to:
# x + y + s1 = 9
# 5x + y + s2 = 11
# x, y, s1, s2 >= 0

# Initial Simplex Tableau
# [ x   y  s1  s2 | RHS ]
# [ 1   1   1   0 |  9 ]  -> Constraint 1
# [ 5   1   0   1 | 11 ]  -> Constraint 2
# [-3  -2   0   0 |  0 ]  -> Objective function (negated coefficients)

tableau = np.array([
    [1, 1, 1, 0, 9],
    [5, 1, 0, 1, 11],
    [-3, -2, 0, 0, 0]
], dtype=float)

print("Initial Simplex Tableau:")
display(tableau)

# Simplex Method Implementation (Manual Pivoting for demonstration)

# Iterate until all values in the last row (excluding the last element) are non-negative
while np.any(tableau[-1, :-1] < 0):
    # Step 1: Find the pivot column (most negative value in the last row)
    pivot_col_index = np.argmin(tableau[-1, :-1])
    print(f"\nPivot Column Index: {pivot_col_index}")

    # Step 2: Find the pivot row (smallest non-negative ratio of RHS to pivot column)
    # Avoid division by zero or negative values in the pivot column
    ratios = []
    for i in range(len(tableau) - 1):
        if tableau[i, pivot_col_index] > 0:
            ratios.append(tableau[i, -1] / tableau[i, pivot_col_index])
        else:
            ratios.append(np.inf) # Use infinity for non-positive divisors

    if np.all(np.isinf(ratios)):
        print("\nProblem is unbounded. No optimal solution.")
        break # Exit if all ratios are infinity

    pivot_row_index = np.argmin(ratios)
    print(f"Ratios: {ratios}")
    print(f"Pivot Row Index: {pivot_row_index}")

    # Step 3: Perform pivoting
    # Divide the pivot row by the pivot element to make the pivot element 1
    pivot_element = tableau[pivot_row_index, pivot_col_index]
    tableau[pivot_row_index, :] /= pivot_element

    # Make all other elements in the pivot column zero by row operations
    for i in range(len(tableau)):
        if i != pivot_row_index:
            factor = tableau[i, pivot_col_index]
            tableau[i, :] -= factor * tableau[pivot_row_index, :]

    print("\nTableau after pivot:")
    display(tableau)

# If the loop finished because all values in the last row are non-negative
if np.all(tableau[-1, :-1] >= 0):
    print("\nOptimal solution reached.")
    # Extract the optimal solution
    # The variables corresponding to columns with a single 1 and rest 0 in the identity matrix part are basic variables
    # Their values are in the RHS column
    # Non-basic variables are 0
    solution = {}
    for j in range(len(tableau[0]) - 1):
        col = tableau[:, j]
        if np.sum(col == 0) == len(col) - 1 and np.sum(col == 1) == 1:
            row_index = np.where(col == 1)[0][0]
            if j == 0:
                solution['x'] = tableau[row_index, -1]
            elif j == 1:
                solution['y'] = tableau[row_index, -1]
            # You can add s1 and s2 to the solution dictionary if needed
            # elif j == 2:
            #     solution['s1'] = tableau[row_index, -1]
            # elif j == 3:
            #     solution['s2'] = tableau[row_index, -1]

    # The maximum value of the objective function is the negated value in the bottom-right corner
    maximum_value = -tableau[-1, -1]
    solution['maximum_f(x, y)'] = maximum_value


    print("\nOptimal Solution:")
    display(solution)
```

## Linear Programming Applications Problem 7

7. An accounting firm has 900 hours of staff time and 276 hours of reviewing time available each week. The firm charges 180 dollars for an audit and 45 dollars for a tax return. Each audit requires 60 hours of staff time and 12 hours of review time. Each tax return requires 10 hours of staff time and 6 hours of review time. What number of audits and tax returns will yield the maximum revenue?

Let us start by stating the function we wish to maximize as well as it the conditions it is subject to. Like so: $r(x) = 180x + 45y$

\begin{cases}
  $60x + 10y \leq 900$ \\
  $12x + 6y \leq 276$ \\
  $x,y \geq 0$
\end{cases}

With our set up complete we shall move onto the next step: finding our corner points. I shall, again, opt to use the Gauss-Jordan method and set up out array.
\begin{array}{cc|r}
1 & 0 & 0 \\ 60 & 10 & 900
\end{array}

Next, we apply the row operation: $-60R_1 + R_2  R_2$.

\begin{array}{cc|r}
1 & 0 & 0 \\ 0 & 10 & 900
\end{array}

Now, we apply: $\frac{1}{10}R_2  R_2$.

\begin{array}{cc|r}
1 & 0 & 0 \\ 0 & 1 & 90
\end{array}

And thus, we have one corner point at (0,90)! Hooray!

This is the visualization of the function $r(x)$ while subject to it's constraints. Let's introduce the simplex method in order to find the just the maximum.

```python
import numpy as np

# Define the coefficients of the objective function and constraints
# Maximize r(x, y) = 180x + 45y
# Subject to:
# 60x + 10y <= 900
# 12x + 6y <= 276
# x >= 0, y >= 0

# Convert to standard form by adding slack variables s1 and s2
# Maximize r(x, y) = 180x + 45y + 0s1 + 0s2
# Subject to:
# 60x + 10y + s1 = 900
# 12x + 6y + s2 = 276
# x, y, s1, s2 >= 0

# Initial Simplex Tableau
# [  x    y   s1   s2 | RHS ]
# [ 60   10    1    0 | 900 ]  -> Constraint 1
# [ 12    6    0    1 | 276 ]  -> Constraint 2
# [-180 -45    0    0 |   0 ]  -> Objective function (negated coefficients)

tableau = np.array([
    [60, 10, 1, 0, 900],
    [12, 6, 0, 1, 276],
    [-180, -45, 0, 0, 0]
], dtype=float)

print("Initial Simplex Tableau:")
display(tableau)

# Simplex Method Implementation (Manual Pivoting for demonstration)

# Step 1: Find the pivot column (most negative value in the last row)
pivot_col_index = np.argmin(tableau[-1, :-1])
print(f"\nPivot Column Index: {pivot_col_index}")

# Step 2: Find the pivot row (smallest non-negative ratio of RHS to pivot column)
# Avoid division by zero or negative values in the pivot column
ratios = []
for i in range(len(tableau) - 1):
    if tableau[i, pivot_col_index] > 0:
        ratios.append(tableau[i, -1] / tableau[i, pivot_col_index])
    else:
        ratios.append(np.inf) # Use infinity for non-positive divisors

pivot_row_index = np.argmin(ratios)
print(f"Ratios: {ratios}")
print(f"Pivot Row Index: {pivot_row_index}")

# Step 3: Perform pivoting
# Divide the pivot row by the pivot element to make the pivot element 1
pivot_element = tableau[pivot_row_index, pivot_col_index]
tableau[pivot_row_index, :] /= pivot_element

# Make all other elements in the pivot column zero by row operations
for i in range(len(tableau)):
    if i != pivot_row_index:
        factor = tableau[i, pivot_col_index]
        tableau[i, :] -= factor * tableau[pivot_row_index, :]

print("\nTableau after first pivot:")
display(tableau)

# Check if optimal solution is reached (no negative values in the last row)
if np.all(tableau[-1, :-1] >= 0):
    print("\nOptimal solution reached.")
    # Extract the optimal solution
    # The variables corresponding to columns with a single 1 and rest 0 in the identity matrix part are basic variables
    # Their values are in the RHS column
    # Non-basic variables are 0
    solution = {}
    for j in range(len(tableau[0]) - 1):
        col = tableau[:, j]
        if np.sum(col == 0) == len(col) - 1 and np.sum(col == 1) == 1:
            row_index = np.where(col == 1)[0][0]
            if j == 0:
                solution['x'] = tableau[row_index, -1]
            elif j == 1:
                solution['y'] = tableau[row_index, -1]
            # Add other variables (s1, s2) if needed
    solution['revenue'] = tableau[-1, -1]

    print("\nOptimal Solution:")
    display(solution)

else:
    print("\nOptimal solution not reached. Further pivoting is required.")
    # In a full implementation, you would repeat steps 1-3 until optimal solution is reached.
    # For this problem, one more pivot might be needed.
    # Let's perform the second pivot manually for demonstration.

    # Step 1: Find the new pivot column
    pivot_col_index = np.argmin(tableau[-1, :-1])
    print(f"\nNew Pivot Column Index: {pivot_col_index}")

    # Step 2: Find the new pivot row
    ratios = []
    for i in range(len(tableau) - 1):
        if tableau[i, pivot_col_index] > 0:
            ratios.append(tableau[i, -1] / tableau[i, pivot_col_index])
        else:
            ratios.append(np.inf)

    pivot_row_index = np.argmin(ratios)
    print(f"New Ratios: {ratios}")
    print(f"New Pivot Row Index: {pivot_row_index}")

    # Step 3: Perform the second pivoting
    pivot_element = tableau[pivot_row_index, pivot_col_index]
    tableau[pivot_row_index, :] /= pivot_element

    for i in range(len(tableau)):
        if i != pivot_row_index:
            factor = tableau[i, pivot_col_index]
            tableau[i, :] -= factor * tableau[pivot_row_index, :]

    print("\nTableau after second pivot:")
    display(tableau)

    # Check for optimality again
    if np.all(tableau[-1, :-1] >= 0):
        print("\nOptimal solution reached.")
        solution = {}
        for j in range(len(tableau[0]) - 1):
            col = tableau[:, j]
            if np.sum(col == 0) == len(col) - 1 and np.sum(col == 1) == 1:
                row_index = np.where(col == 1)[0][0]
                if j == 0:
                    solution['x'] = tableau[row_index, -1]
                elif j == 1:
                    solution['y'] = tableau[row_index, -1]
                # Add other variables (s1, s2) if needed
        solution['revenue'] = tableau[-1, -1]

        print("\nOptimal Solution:")
        display(solution)
    else:
        print("\nOptimal solution not reached after two pivots. More iterations might be needed.")
```
