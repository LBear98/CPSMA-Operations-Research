Perhaps the greatest strength of python is iterating over lists.  Use the difference formula of calculus to show that the derivative of $\ln x$ at 1 is 1.

In other words show that as $h$ gets smaller and smaller $\frac{\ln(1+h)-\ln(1)}h$ approaches 1.  Mathematically we write this as
$$
\lim_{h\to 0} \frac{\ln(1+h)-\ln(1)}h = 1
$$


Use the list provided below.

Extra challenge:  Use the `sympy` package to compute the limit and derivative.

```python
h = [1,0.1,0.001,0.001,0.0001,0.0000000001]
```

```python
from sympy import limit, symbols, log, diff

# Define the symbol x and h
x, h = symbols('x h')

# Define the function
f = log(x)

# Compute the derivative of log(x)
derivative_f = diff(f, x)
print("The derivative of ln(x) is:", derivative_f)

# Compute the limit of (ln(1+h) - ln(1)) / h as h approaches 0
limit_expr = (log(1 + h) - log(1)) / h
limit_result = limit(limit_expr, h, 0)
print("The limit of (ln(1+h) - ln(1)) / h as h approaches 0 is:", limit_result)
```

## Truth of Summation

Verify the following summation formula using python for all $n$ where $n\in\mathbb{Z}$ and $n\leq 1000$.

$$
\sum_{i = 0}^n i^3 = \left(\frac{n(n+1)}{2}\right)^2
$$

Extra challenge:  Reduce the output to a single true or false.  Do not attempt until original has been verified with a clean code.

```python
for n in range(1001):
    sum_of_cubes = sum(i**3 for i in range(n + 1))
    formula_result = (n * (n + 1) // 2)**2
    if sum_of_cubes != formula_result:
        print(f"Formula is false for n = {n}")
        break
else:
    print("Formula is true for all n <= 1000")
```

Alrighty, we have achieved our first positive result, now let's break it down to a single cell.

```python
is_true = True
for n in range(1001):
    sum_of_cubes = sum(i**3 for i in range(n + 1))
    formula_result = (n * (n + 1) // 2)**2
    if sum_of_cubes != formula_result:
        is_true = False
        break
print(is_true)
```

## Factorials

A factorial is the product of all previous numbers and is expressed with an exclamation $!$.  Thus
$$
3! = 3\cdot 2\cdot 1 = 6
$$

and
$$
n! = n(n-1)(n-2)\cdots3\cdot2\cdot1
$$

How many zeros does $1000!$ end with?  Please do not count them in the number below as the precision may be off!

```python
import math

math.factorial(1000)
```

```python
def count_trailing_zeros(n):
    count = 0
    i = 5
    while n // i >= 1:
        count += n // i
        i *= 5
    return count

n = 1000
zeros = count_trailing_zeros(n)
print(f"The number of zeros at the end of {n}! is: {zeros}")
```

## Handshakes

Two soccer teams have just finished their game.  The two teams of 11 all shake hands and the winning team also shakes hand with each other.  How many handshakes have occured?

The amount of handshakes that occur is $132$. This can be express as either $11*12$ or $(11*11) + 11$ given the winning team also shakes hands with each other. It could even be argued that 181 handshakes occur since winning team would shake hands with itself at a ratio of 5 to 6. Thus, $11*11 + 5*6 + 6*5$ could also be used to express the number of handshakes.
